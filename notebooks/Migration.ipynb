# GOLDILOX Insights — Migration Setup

This notebook prepares your account for migrating GOLDILOX Insights to a new Snowflake Marketplace listing.

**Run this notebook as ACCOUNTADMIN** (or a role with sufficient privileges).

### What this notebook does

1. Creates a backup schema in your existing workspace database
2. Grants the app read/write access so it can backup and restore data

### After running this notebook

Open the app and go to **Settings → Migration** to back up your data before uninstalling.

---
## Configuration

Update `APP_NAME` below if you renamed the app during installation. Otherwise, run as-is.

```
-- Default app name (update only if you renamed the app during installation)
SET APP_NAME = 'GOLDILOX_INSIGHTS';
```

---
## 1. Create Backup Schema

Creates a schema in your existing `GOLDILOX_INSIGHTS_CLIENT_WORKSPACE` database. This is where the app will store backup data. The schema **survives the app uninstall**.

```
CREATE SCHEMA IF NOT EXISTS GOLDILOX_INSIGHTS_CLIENT_WORKSPACE.MIGRATION_BACKUP;
```

---
## 2. Grant App Access

Grants the app read/write access to the backup schema so it can create backup tables and later read them for import.

> **After reinstalling the new app**, re-run this cell to grant the new app access.

```
GRANT USAGE ON DATABASE GOLDILOX_INSIGHTS_CLIENT_WORKSPACE TO APPLICATION IDENTIFIER($APP_NAME);
GRANT USAGE ON SCHEMA GOLDILOX_INSIGHTS_CLIENT_WORKSPACE.MIGRATION_BACKUP TO APPLICATION IDENTIFIER($APP_NAME);
GRANT CREATE TABLE ON SCHEMA GOLDILOX_INSIGHTS_CLIENT_WORKSPACE.MIGRATION_BACKUP TO APPLICATION IDENTIFIER($APP_NAME);
GRANT SELECT ON ALL TABLES IN SCHEMA GOLDILOX_INSIGHTS_CLIENT_WORKSPACE.MIGRATION_BACKUP TO APPLICATION IDENTIFIER($APP_NAME);
```

---
## Next Steps

1. **Open the app → Settings → Migration** and click **Back Up Data**
2. **Uninstall the old app:** `DROP APPLICATION GOLDILOX_INSIGHTS;` (or your custom app name)
3. **Install the new app** from the new Snowflake Marketplace listing
4. **Run the `Setup_App_Permissions` notebook** to re-grant warehouse MONITOR and database access
5. **Re-run Step 2 above** to grant the new app access to the backup schema
6. **Open the new app → Settings → Migration** and click **Import Data**

> The `GOLDILOX_INSIGHTS_CLIENT_WORKSPACE` database is customer-owned and will **not** be affected by the app uninstall.
